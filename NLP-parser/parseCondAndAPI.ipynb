```python
testStr = "The dragon attacks the player"

def removeStopWords(inputWordToken):
    stop_words = set(stopwords.words('english'))
  
    #word_tokens = word_tokenize(inputStr)
    
    
    # converts the words in word_tokens to lower case and then checks whether 
    #they are present in stop_words or not
    filtered_sentence = [w for w in inputWordToken if not w.lower() in stop_words]
    #with no lower case conversion
    filtered_sentence = []

    for w in inputWordToken:
        if w not in stop_words:
            filtered_sentence.append(w)
    return filtered_sentence

def splitSpeechTags(tagList):
    nounList = []
    verbList = []
    for i,tag in enumerate(tagList):
        if tag[1] == "NN":
            nounList.append(tag)
        elif tag[1] == "VBZ" or tag[1] == "NNS":
            verbList.append(tag)
    return nounList, verbList
```

```python
def getSpeechTags(inputStr):
    stemmer = SnowballStemmer("english")

    testStrTok = nltk.word_tokenize(inputStr.lower())
    testStrTok = removeStopWords(testStrTok)
    tagList = nltk.pos_tag(testStrTok)
    nounList, verbList = splitSpeechTags(tagList)
    
    return nounList, verbList


nounList, verbList = getSpeechTags(testStr)
print("Noun: ", nounList)
print("Verb: ", verbList)
```

```python

# THIS WILL BE THE MAIN FUNCTION YOU WILL CALL
def getCondAndActionTag(inputStr): 

    #print(testStr,'\n\n\n\n')

    # THEN is a key word (for splitting if statements)
    thenList = inputStr.split('then')

    assert len(thenList) == 2

    condTag = getConditionTag(thenList[0])
    #print('Condition Tag: ', condTag)

    actionList = []
    for action in thenList[1].split('and'):
        actionTag = getActionTag(action)
        #print('Action Tag: ', actionTag)
        actionList.append(actionTag)
    return condTag,actionList
```
